```python
# Importando para checar a versão
import matplotlib
import sklearn

# Importando as bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

print(f"Versão pandas: {pd.__version__}")
print(f"Versão matplotlib: {matplotlib.__version__}")
print(f"Versão seaborn: {sns.__version__}")
print(f"Versão sklearn: {sklearn.__version__}")
```

### Quadrático

```python
# Criando um dicionário com os dados
dic_data = {1:[23, 88],
          2:[25, 74],
          3:[26, 60],
          4:[28, 44],
          5:[30, 46],
          6:[32, 56],
          7:[35, 67],
          8:[36, 76],
          9:[38, 90]}

# Transformando o dicionário em dataframe
df = pd.DataFrame(dic_data, index = ['Age', 'hour_price']).T

# Checando as primeiras linhas
df.head()
```

```python
# Checando a relação
sns.scatterplot(data = df, x = 'Age', y = 'hour_price')
```

```python
# Inicializando o modelo
lr = LinearRegression()
```

```python
# Dividindo as colunas em independente e dependente
x = df['Age'].values.reshape(-1, 1)
y = df['hour_price'].values.reshape(-1, 1)

# Treinando o modelo e realizando predições
lr.fit(x, y)
predict = lr.predict(x)

# Plotando o gráfico
sns.scatterplot(data = df, x = 'Age', y = 'hour_price')
plt.plot(df['Age'], predict, color = 'r')
```

```python
# Criando um transformador polinomial de ordem quadrática
poly_t = PolynomialFeatures(degree=2)

# Transformando os dados e criando o modelo 
x_poly = poly_t.fit_transform(df['Age'].values.reshape(-1, 1))
lr.fit(x_poly, y)
predict_poly = lr.predict(x_poly)

# Plotando o gráfico
sns.scatterplot(data = df, x = 'Age', y = 'hour_price')
plt.plot(df['Age'], predict_poly, color = 'r')
```

### Testando diferentes tipos de Grau

```python
# Definindo o grau
grau = 2

# Instanciando os métodos
lr = LinearRegression()
poly_t = PolynomialFeatures(degree=grau)

# Prevendo os resultados
x_poly = poly_t.fit_transform(df['Age'].values.reshape(-1, 1))
lr.fit(x_poly, y)
predict_poly = lr.predict(x_poly)

# Plotando o gráfico
sns.scatterplot(data = df, x = 'Age', y = 'hour_price', label = f'grau: {grau}')
plt.plot(df['Age'], predict_poly, color = 'r')
plt.legend(loc = (1.03, 0.90))
```
